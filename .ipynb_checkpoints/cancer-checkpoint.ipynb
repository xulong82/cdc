```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "/Users/xulongwang/Git/cdc/cancer.xlsx"
dat = pd.read_excel(file, sheet_name = "Lifetime Probability of Develop", header = 0, skiprows = 2, nrows = 24)
dat
```

```python
dat = dat.drop(['Unnamed: 4', 'Unnamed: 9'], axis = 1)
dat.head()
```

```python
dat1 = dat.iloc[:, 0:5]
dat1.columns = ['Site', 'LT%dev', '1in?dev', 'LT%die', '1in?die']
dat1.loc[:,'LT%dev'] = pd.to_numeric(dat1['LT%dev'], errors = 'coerce')
dat1 = dat1.assign(Sex = 'M')
dat1
```

```python
dat2 = dat.drop(dat.columns[1:5], axis = 1)
dat2.columns = ['Site', 'LT%dev', '1in?dev', 'LT%die', '1in?die']
dat2.loc[:,'LT%dev'] = pd.to_numeric(dat2['LT%dev'], errors = 'coerce')
dat2 = dat2.assign(Sex = 'F')
dat2
```

```python
dat3 = dat1.sort_values(by=['LT%dev'])
dat3.dropna(subset = ['LT%dev'], inplace=True)

plt.rcParams["figure.figsize"] = (15,8)
plt.barh(dat3['Site'], dat3['LT%dev'], align='center')
plt.xlabel('Lifetime Probability of Development')
plt.title('Cancer Prevelence in US 2018: Male')
for x,y in enumerate(dat3['LT%dev']): plt.text(y, x, str(y)+'%', color='red')
plt.show()
```

```python
dat4 = dat2.sort_values(by=['LT%dev'])
dat4.dropna(subset = ['LT%dev'], inplace=True)

plt.barh(dat4['Site'], dat4['LT%dev'], align='center')
plt.xlabel('Lifetime Probability of Development')
plt.title('Cancer Prevelence in US 2018: Female')
for x,y in enumerate(dat4['LT%dev']): plt.text(y, x, str(y)+'%', color='red')
plt.show()
```

```python
dat5 = dat1.append(dat2, ignore_index=True)
dat5 = dat5.loc[dat5['LT%dev'] > 2, ]
dat5 = dat5.sort_values(by = ['LT%dev'], ascending = False)
dat5.reset_index(inplace = True, drop = True)
dat5
```
